```python
import pandas as pd

df = pd.read_csv("../data/listings.csv.gz", compression="gzip")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 5))

sns.histplot(df, x="host_id", ax=ax)
plt.savefig('number_of_properties_per_host.png')
plt.show()
```

```python
sns.pairplot(df.sample(frac=0.2), vars=[
    'host_listings_count', 'maximum_nights_avg_ntm', 'minimum_nights_avg_ntm', 'number_of_reviews', 'review_scores_rating'
])

plt.show()
```
