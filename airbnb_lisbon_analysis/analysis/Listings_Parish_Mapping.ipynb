```python
import pandas as pd

df = pd.read_csv("../data/listings.csv.gz", compression="gzip")
reviews = pd.read_csv("../data/reviews.csv.gz", compression="gzip")
review_languages = pd.read_csv("../data/review_languages.csv.gz", compression="gzip")
```

```python
import geopandas
from shapely.geometry import Point

df_parishes = geopandas.read_file("../data/lisbon_parishes.geojson")
df["geometry"] = df.apply(lambda row: Point(row["longitude"], row["latitude"]), axis=1)

df_listings = geopandas.GeoDataFrame(df, crs="EPSG:4326", geometry="geometry")
```

```python
df_listings.head()
```

```python
from collections import defaultdict

listings = []
listings_per_parish = defaultdict(list)

for _, parish in df_parishes.iterrows():
    filtered_listings = df_listings[df_listings["geometry"].within(parish.geometry)]
    listings.append(filtered_listings)
    listings_per_parish[parish.id].extend(filtered_listings.id.tolist())

df_listings = pd.concat(listings)
df_listings = df_listings[df_listings["number_of_reviews"] > 0]
```

For every parish, we get all the listings. We calculate the majority language per every listing. 

```python
from collections import Counter

listing_language = {}

for _, listing in df_listings.iterrows():
    review_ids_per_listing = reviews[reviews["listing_id"] == listing["id"]]["id"].tolist()
    languages = review_languages[review_languages["id"].isin(review_ids_per_listing)]["language"].tolist()
    if len(languages):
        language_counter = Counter(languages)
        listing_language[listing["id"]] = language_counter.most_common(1)[0][0]
```

Then the majority language per parish:

```python
parish_language = {}
parish_count = defaultdict(0)

for parish, listing_ids in listings_per_parish.items():
    languages = [listing_language[listing] for listing in listing_ids if listing in listing_language]
    language_counter = Counter(languages)
    
    if len(languages):
        parish_language[parish] = language_counter.most_common(1)[0][0]
```

```python
import csv

with open("../data/parish_data.csv", "wt") as fp:
    writer = csv.writer(fp)
    writer.writerow(("parish_id", "language"))
    writer.writerows(list(parish_language.items()))
```
